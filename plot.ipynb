```python
import matplotlib.pyplot as plt
```

**Case 1**

```python
thread_counts = ["1", "2", "3", "4"]
serial_avg_time = 14.76
mutex_avg_times = [14.8, 17.57, 17.54, 21.54]
rw_avg_times = [21.75, 14.3, 9.14, 7.28]
```

```python
plt.plot(thread_counts, mutex_avg_times, color='r', label='One mutex for entire list', marker="o", markeredgecolor="r", markerfacecolor="r")
plt.plot(thread_counts, rw_avg_times, color='g', label='Read-Write lock', marker="o", markeredgecolor="g", markerfacecolor="g")
plt.plot(thread_counts[0], serial_avg_time, color='b', label='Serial', marker="o", markeredgecolor="b", markerfacecolor="b")

plt.xlabel("# Threads")
plt.ylabel("Average Execution Time (ms)")
plt.title("Case 1")

plt.legend()
plt.show()
```
